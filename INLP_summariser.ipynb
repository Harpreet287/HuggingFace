```python
from transformers import pipeline
import torch
```

```python
summarizer = pipeline(task = 'summarization', model ="facebook/bart-large-cnn", torch_dtype = torch.bfloat16)
```

```python
text = """Tom and Jerry is an American animated media franchise and series of comedy short
          films created in 1940 by William Hanna and Joseph Barbera.
          Best known for its 161 theatrical short films by Metro-Goldwyn-Mayer,
          the series centers on the rivalry between the titular
          characters of a cat named Tom and a mouse named Jerry.
          Many shorts also feature several recurring characters.
"""
```

```python
summary = summarizer(text, min_length=5, max_length=15)
```
