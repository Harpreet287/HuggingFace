## Building a simple chatbot pipeline using HuggingFace Transformers Library, that can talk back and remember your chats


```python
from transformers import pipeline
```

```python
chatbot = pipeline(task = 'conversational', model = 'facebook/blenderbot-400M-distill')
```

```python
import torch
```

```python
user_message = """How to talk to someone for the first time"""
```

```python
from transformers import Conversation
```

```python
conversation = Conversation(user_message)
conversation
```

```python
conversation = chatbot(conversation) # the `chatbot` is the instance i made
conversation
```

```python
# inorder for ask successively
conversation.add_message({"role":"user", "content":"""How to order a drink?"""})
print(conversation)
```

```python
print(chatbot(conversation))
```

```python
conversation.add_message({"role":"user", "content":"""How to order a waiter?"""})
print(conversation,chatbot(conversation))
```
